# __Rekomendasi Film__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
dataFilm = pd.read_csv('movies.csv')
dataFilm.head()
```

```python
dataRating = pd.read_csv('ratings.csv')
dataRating = dataRating[['userId', 'movieId', 'rating']]
dataRating
```

```python
cv = CountVectorizer()
cm = cv.fit_transform(dataFilm['genres'])

print(cv.get_feature_names())
print(cm.toarray())
```

```python
similarityScore = cosine_similarity(cm.toarray())
similarityScore
```

```python
jokoFav = dataFilm[dataFilm['title'] == 'Superman vs. The Elite (2012)'].index[0]
jokoFav
```

```python
similarMovie = list(enumerate(similarityScore[jokoFav]))
similarMovie = sorted(similarMovie, key=lambda x: x[1], reverse=True)
# similarMovie
```

### 5 Rekomendasi Film Berdasarkan Genre Film kesukaan Joko

```python
dfSim = []
for i in similarMovie[:6]:
    dfSim.append(dataFilm.iloc[i[0]])
dfSim = pd.DataFrame(dfSim)
dfSim = dfSim[dfSim['title']!= 'Superman vs. The Elite (2012)']
dfSim
```

## Collaborative Filtering

```python
ratingWidodo = [(dataFilm[dataFilm['title'] == 'Being Flynn (2012)'].movieId.values[0], 7)]
ratingWidodo
```

```python
df = dataRating.pivot_table(
    index = 'userId',
    columns = 'movieId'
)
df = df.replace([np.NaN, -1], 0) # ganti yg tidak kasih rating dan blm nntn
df
```

```python
dfCorr = df.corr().loc['rating']
dfCorr = dfCorr.fillna(0) # yang hasil correlasi NaN karena 0/0 = NaN, diganti menjadi 0
dfCorr
```

```python
# dfCorr.loc[95816]
```

```python
scoreSimilar = pd.DataFrame()
for movieId, rating in ratingWidodo:
    score = dfCorr.loc[movieId] * (rating/10)
    score = score.sort_values(ascending=False)
    scoreSimilar = scoreSimilar.append(score, ignore_index=True)
scoreSimilar
```

```python
dfSkor = pd.DataFrame(scoreSimilar.sum().sort_values(ascending=False))
dfSkor
```
